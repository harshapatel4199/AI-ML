## build a linear regression

```python
import matplotlib.pyplot as plt
import numpy as np

# How Linear Regression works:
# y = mx + b

# let take,
# x = 2*(random number)
# y = 4 + 3x + (random number)

# now,
# y = mx+b

np.random.seed(42)
x = 2 * np.random.rand(100,1)
y = 4 + 3 * x + np.random.randn(100,1)

m = 0 
b = 0 

learning_rate = 0.1
iterantions = 1000
n = len(x)

for i in range(iterantions):
    y_pred = m * x + b
    error = y_pred - y

    dm = (2/n) * np.sum(error * x)
    db = (2/n) * np.sum(error)

    m -= learning_rate*dm
    b -= learning_rate*db

    if i % 100 == 0:
        loss = np.mean(error**2)
        print(f"Iteration {i}: Loss = {loss:.4f}")

print(f"Final Equation : y = {m:.2f}x + {b:.2f}")

plt.scatter(x,y, color='blue', label='Actual Data')
plt.plot(x, m * x + b, color ='red', label='Predicted Line')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
